## Missing values

```python
cols_with_na = [col 
                for col in data.columns 
                if data[col].isna().sum()>0]
cols_with_na
```

### Relationship between missing values and Sale price

```python
def analyse_na_value(df, col):
    df = df.copy()
    
    df[col] = np.where(df[col].isna(), 'missing', 'not missing')
    df.groupby(col)['SalePrice'].median().plot.bar()
    
    plt.title(col)
    plt.show()

for col in cols_with_na:
    analyse_na_value(data, col)
```

#### Temporal variables

```python
year_cols = [col for col in numeric_cols if 'Yr' in col or 'Year' in col]
data[year_cols].head()
```

```python
for col in year_cols:
    print(data[col].unique())
```

```python
median_price_vs_yr_sold = data.groupby('YrSold')['SalePrice'].median()
print(median_price_vs_yr_sold)

median_price_vs_yr_sold.plot()
plt.ylabel("Median House Price")
plt.title("Change in House price with years")
```

```python
def analyse_year_cols(df, col):
    df = df.copy()
    df[col] = df['YrSold'] - df[col]
    
    plt.scatter(df[col], df['SalePrice'])
    plt.ylabel('SalePrice')
    plt.xlabel(col)
    plt.show()
    
for col in year_cols:
    if col != 'YrSold':
        analyse_year_cols(data, col)
```

### Discrete variables

```python
ignore_cols = year_cols + ['Id']
discrete_cols = [col 
                 for col in numeric_cols 
                 if len(data[col].unique()) < 20 
                 and col not in ignore_cols
                ]
discrete_cols
```

```python
data[discrete_cols]
```

```python
def analyse_discrete(df, col):
    df = df.copy()
    df.groupby(col)['SalePrice'].median().plot.bar()
    
    plt.title(f"{col} vs. SalePrice")
    plt.ylabel("SalePrice")
    plt.show()

for col in discrete_cols:
    analyse_discrete(data, col)
```

#### Continuous variables

```python
continuous_cols = [col 
                   for col in numeric_cols 
                   if col not in discrete_cols + year_cols + ["Id"]]
```

```python
continuous_cols
```

```python
data[continuous_cols].head()
```

```python
def analyse_continuous(df, col):
    df= df.copy()
    
    df[col].hist(bins=20)
    
    plt.title(f"{col} vs No. of houses")
    plt.ylabel("Number of houses")
    plt.xlabel(col)
    plt.show()
```

```python
for col in continuous_cols:
    analyse_continuous(data, col)
```

We see that all of the above variables, are not normally distributed, including the target variable 'SalePrice'. For linear models to perform best, we need to account for non-Gaussian distributions. We will transform our variables in the next lecture / video, during our feature engineering section.

Let's also evaluate here if a log transformation renders the variables more Gaussian looking:

```python
def analyse_log_transformed_continuous(df, col):
    df = df.copy()
    
    # log does not take negative
    if any([x<=0 for x in data[col].unique()]):
        print(f"{col} contains non positive values.. Skipping")
        return
    df[col] = np.log(df[col])
    df[col].hist(bins=20)
    
    plt.ylabel("No. of houses")
    plt.xlabel(col)
    plt.title(f"{col} vs No.of houses")
    plt.show()
```

```python
for col in continuous_cols:
    analyse_log_transformed_continuous(data, col)
```

We get a better spread of values for most variables when we use the logarithmic transformation. This engineering step will most likely add performance value to our final model.

Let's explore the relationship between the house price and the transformed variables
with more detail

```python
def transform_analyse_continuous(df, col):
    df = df.copy()
    
    # log does not take negative
    if any([x<=0 for x in data[col].unique()]):
        print(f"{col} contains non positive values.. Skipping")
        return

    df[col] = np.log(df[col])
    df['SalePrice'] = np.log(df['SalePrice'])
    
    plt.scatter(df[col], df['SalePrice'])
    plt.ylabel("No. of houses")
    plt.xlabel(col)
    plt.title(f"{col} vs No.of houses")
    plt.show()

for col in continuous_cols:
    if col != 'SalePrice':
        transform_analyse_continuous(data, col)
```

#### Outliers

The majority of the continuous variables seem to contain outliers. Outliers tend to affect the performance of linear model. So it is worth spending some time understanding if removing outliers will add performance value to our final machine learning model.

#### Rare Labels

```python
def analyse_rare_labels(df, col, threshold_percentage):
    df = df.copy()
    tmp = df.groupby(col)['SalePrice'].count() / len(df)
    return tmp[tmp < threshold_percentage]
```

```python
for c in categorical_cols:
    print(analyse_rare_labels(data, c, 0.01))
```

Some of the categorical variables show multiple labels that are present in less than 1% of the houses. We will engineer these variables in our next video. **Labels that are under-represented in the dataset tend to cause over-fitting of machine learning models. That is why we want to remove them.**

```python
for c in categorical_cols:
    analyse_discrete(data, c)
```
